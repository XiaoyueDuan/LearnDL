````python
from keras.applications.vgg16 import VGG16
import numpy as np
from numpy.random import permutation
from keras.layers.core import Dense
from keras.models import Model
from keras.optimizers import RMSprop, Adam, SGD

from matplotlib import pyplot as plt

%matplotlib inline
````

````python
model=VGG16(include_top=True, weights='imagenet')
````

## Action PlanÂ¶
1. Modify the last layer of VGG16
2. Train the model and calculate the network ability by validate set
3. Predict on test dataset
4. Generate files to submit on Kaggle

### 1. Modify the last layer of VGG16

````python
model.layers.pop()
x=model.output
x=Dense(2, activation='softmax')(x)
new_model=Model(input=model.input, outputs=x)
````

````python
for layer in model.layers:
    layer.trainable = False
````

````python
new_model.summary()
````

````python
sgd = SGD(lr=1e-3, decay=1e-6, momentum=0.9, nesterov=True)
new_model.compile(optimizer=sgd, loss='categorical_crossentropy', metrics=['accuracy'])
````

### 2. Train the model and calculate the network ability by validate set

````python
from keras.preprocessing import image

data_home_dir='..\\dogs_cats_dataset\\dogscats\\'

gen=image.ImageDataGenerator()
image_shape=(224,224)
train_batch=gen.flow_from_directory(data_home_dir+'train', batch_size=16, target_size=image_shape)
valid_batch=gen.flow_from_directory(data_home_dir+'valid', batch_size=16, target_size=image_shape, shuffle=False)
````

````python
new_model.fit_generator(train_batch, steps_per_epoch=np.ceil(train_batch.n/train_batch.batch_size).astype(int),
                    validation_data=valid_batch, validation_steps=np.ceil(valid_batch.n/valid_batch.batch_size).astype(int), epochs=1)
````

### Save training weight

````python
weights_folder='../weights/'
new_model.save_weights(weights_folder+'finetune_vgg16.h5')
````

### 4. Generate files to submit on Kaggle
Because the order of files in test_batch is:
```
    1.jpg
    10.jpg
    100.jpg
    1000.jpg
    1001.jpg
    ...
```
We need to extract the number in its name so that sort them in the following order:
```
    1.jpg
    2.jpg
    3.jpg
    4.jpg
    5.jpg
    ...
```
We implement the function by **Regular Expression**.

````python
is_dog=prob[:,1]
````

````python
import re

def find_num(pattern='\d*(?=\.jpg$)'):
    def f(str_to_be_matched):
        return re.findall(pattern, str_to_be_matched)[0]
    return f

index=list(map(find_num('\d*(?=\.jpg$)'), test_batch.filenames))
index=list(map(int, index))
````

````python
index=np.array(index)
is_dog=is_dog[index.argsort()]
````

````python
ids=np.arange(len(is_dog)).reshape((-1,1))+1
sub=np.concatenate([ids, is_dog.reshape(-1,1)],axis=1)
````

````python
submission_file_name = 'real_trained_right_order.csv'
np.savetxt(submission_file_name, sub, fmt='%d,%.5f', header='id,label', comments='')
````

## 5. Visualize by validate datasets

````python
valid_batch=gen.flow_from_directory(data_home_dir+'valid', batch_size=16, target_size=image_shape, shuffle=False) # close shuffle
expected_labels=valid_batch.classes
filenames=valid_batch.filenames
````

````python
is_dog_validate=new_model.predict_generator(valid_batch, np.ceil(valid_batch.n/valid_batch.batch_size).astype(int),verbose=1)
````

````python
our_predictions = is_dog_validate[:,0]
our_labels = np.round(1-our_predictions)
our_predictions[:10]
````

````python
def plots(ims, figsize=(12,6), rows=1, interp=False, titles=None):
    if type(ims[0]) is np.ndarray:
        ims = np.array(ims).astype(np.uint8)
        if (ims.shape[-1] != 3):
            ims = ims.transpose((0,2,3,1))
    f = plt.figure(figsize=figsize)
    cols = len(ims)//rows if len(ims) % 2 == 0 else len(ims)//rows + 1
    for i in range(len(ims)):
        sp = f.add_subplot(rows, cols, i+1)
        sp.axis('Off')
        if titles is not None:
            sp.set_title(titles[i], fontsize=16)
        plt.imshow(ims[i], interpolation=None if interp else 'none')
````

````python
def plots_idx(idx, titles=None):
    plots([image.load_img(data_home_dir+'valid/'+ filenames[i]) for i in idx], titles=titles)
````

````python
correct = np.where(our_labels==expected_labels)[0]
correct.shape
````

````python
n_view=4
````

````python
# 1. A few correct labels at random
print("Found %d correct labels" % len(correct))
idx = permutation(correct)[:n_view]
plots_idx(idx, our_predictions[idx])
````

````python
#2. A few incorrect labels at random
incorrect = np.where(our_labels!=expected_labels)[0]
print("Found %d incorrect labels" % len(incorrect))
idx = permutation(incorrect)[:n_view]
plots_idx(idx, our_predictions[idx])
````

````python
def plot_confusion_matrix(cm, classes, normalize=False, title='Confusion matrix', cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    (This function is copied from the scikit docs.)
    """
    plt.figure()
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    print(cm)
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j], horizontalalignment="center", color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
````
